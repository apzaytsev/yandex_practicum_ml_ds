# Анализ рынка компьютерных игр интернет-магазина «Стримчик»

Нам предоставлены исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

<b>Целью проекта</b> является выявление закономерностей, влияющих на успешность игр, для определения перспективных направлений в будущем.

<b>Описание данных:</b>
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

<b>План действий:</b>
1. Изучение общей информации о загруженном датафрейме.
2. Подготовка данных:
    - Приведение к нижнему регистру названий столбцов.
    - Преобразование данных в нужные типы.
    - Заполнение пропущенных значений там, где это возможно
    - Посчет суммарных продаж во всех регионах.
3. Исследовательский анализ данных.
    - Анализ количества игр в разные годы.
    - Анализ продаж по платформам.
    - Определим актуальный период для дальнейшего анализа. Данные за предыдущие годы не будут учитываться.
    - Определение платформ с наибольшими суммарными продажами и построение распределения по годам.
    - Построение графика «ящик с усами» по глобальным продажам игр в разбивке по платформам.
    - Определение влияния на продажи внутри одной популярной платформы отзывы пользователей и критиков.
    - Общее распределение игр по жанрам.
4. Составление портретов пользователя каждого региона. Определение для пользователя каждого региона (NA, EU, JP):
    - Самые популярные платформы.
    - Самые популярные жанры.
    - Влияние рейтинга ESRB на продажи в отдельном регионе.
5. Проверка гипотез.
    - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
    - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
6. Общий вывод.

### Шаг 2. Подготовка данных

#### Заменим названия столбцов (приведем к нижнему регистру)

```python
games.columns = games.columns.str.lower()
```

Кроме названий столбцов приведем к нижнему регистру и названия игр для поиска неявных дубликатов по некоторым столбцам.

```python
games['name'] = games['name'].str.lower()
```

```python
games[games.duplicated(subset=['name', 'platform', 'year_of_release'], keep=False)]
```

Обнаружен неявный дубликат, удалим строку с индексом 16230. Со строками с пропусками с названии игр разберемся при обработке пропусков в п. 2.3.

```python
games = games.drop(index=16230)
```

#### Преобразуем данные в нужные типы

Преобразуем столбцы <b>year_of_release</b> и <b>critic_score</b> из <b>float</b> в <b>integer</b>

```python
games = games.astype({'year_of_release': 'Int64', 'critic_score': 'Int64'})
```

Столбец оценки пользователей преобразуем после оценки значений 'tbd' (to be determined = будет определено).

#### Обработаем пропуски

```python
games.isnull().sum()
```

```python
games[games['name'].isna()]
```

Данные в указанном срезе смысловой нагрузки не несут, восстановлению не подлежат. Удалим строки с пропусками в столбцах <b>name</b> и <b>genre</b>. 

```python
games = games.dropna(subset = ['name'])
```

```python
games[games['year_of_release'].isna()].shape[0] / games.shape[0]
```

Далее возникает дилемма, что же делать с пропусками в годах выпуска? Можно было бы для начала выявить название игр с датами, но, к примеру, большинство спортивных симуляторов выпускаются в предыдущем календарном году.

💡Что можно сделать?
- ✅ полностью удалить пропуски, поскольку срез составляет всего лишь 1.6% от всего датасета (слишком просто);
- ⛔ для 100% достоверной исходной выборки загуглить каждую игру и год ее выпуска и заполнить пропущенные значения (слишком долго, но к этому варианту всегда можно вернуться);
- ⛔ выбрать какой-то средний вариант, заполнив пропуски по алгоритму и ограничив наиболее значимые игры исходя из совокупной выручки во всех регионах вручную (кажется, оптимально, но нужно выбрать подходящее количество проданных копий), используя словарь.

```python
sorted(games[games['year_of_release'].isna()].query('na_sales + eu_sales + jp_sales + other_sales > 0.5')['name'].unique())
```

Вручную создадим словарь, в котором укажем год релиза игр, суммарные продажи которых превысили 500 тыс. копий. Остальные строки с пропусками в столбце <b>year_of_release</b> удалим.

```python
na_list = {'advance wars: days of ruin': 2008, 
'adventure': 1980, 
'air-sea battle': 1977, 
'combat': 1977,
'donkey kong land iii': 1997,
'fishing derby': 1980,
"frogger's adventures: temple of the frog": 2001,
'legacy of kain: soul reaver': 1999,
'monster hunter 2': 2006, 
'rock band': 2007,
'suikoden iii': 2002,
'test drive unlimited 2': 2011,
'triple play 99': 1998, 
'warioware: twisted!': 2004,
'wwe smackdown vs. raw 2006': 2005,
'yakuza 4': 2010}
```

```python
for i in na_list:
    games.loc[games['name']== i,'year_of_release'] = na_list[i]
```

```python
games = games.dropna(subset = ['year_of_release'])
```

```python
games.info()
```

```python
games['user_score'].sort_values().unique()
```

```python
games['user_score'].loc[games['user_score'] == 'tbd'] = np.nan
```

Теперь преобразуем столбец оценки пользователей в тип данных float.

```python
games = games.astype({'user_score': 'float64'})
```

Обработаем пропуски заглушками.

```python
games['rating'] = games['rating'].fillna('unknown')
games['critic_score'] = games['critic_score'].fillna(-1)
games['user_score'] = games['user_score'].fillna(-1)
```

```python
games.info()
```

#### Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.

### Шаг 5. Проверим гипотезы

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

- H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- H1: Средние пользовательские рейтинги платформ Xbox One и PC различны.

Для проверки гипотезы уберем заглушки -1

```python
user_scores = games_actual.query('user_score >= 0')
```

Использовать будем тест Стьюдента

```python
alpha = 0.05
results = st.ttest_ind(user_scores.query('platform == "XOne"')['user_score'], 
                       user_scores.query('platform == "PC"')['user_score'], equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вывод: с большой долей вероятности средние пользовательские рейтинги платформ XOne и PC не имеют статистически значимых отличий.

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

- H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
- H1: Средние пользовательские рейтинги жанров Action и Sports различны.

```python
alpha = 0.05
results = st.ttest_ind(user_scores.query('genre == "Action"')['user_score'], 
                       user_scores.query('genre == "Sports"')['user_score'], equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вывод: средние пользовательские рейтинги жанров Action и Sports различны.
