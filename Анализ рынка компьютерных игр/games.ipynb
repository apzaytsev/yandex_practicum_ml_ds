```python
def create_plot(platform_list, ticks, platform_title):
    for i in platform_list:
        games[games['platform']== i].groupby('year_of_release')['total_sales'].sum().plot(kind='line', figsize=(15,8))
    mplcyberpunk.add_gradient_fill(alpha_gradientglow=0.4)
    plt.legend(platform_list)
    plt.xticks(ticks=ticks)
    plt.xlabel('')
    plt.ylabel(' , .')
    plt.title('  ' + platform_title + '   ')
    plt.show()
```

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']
def create_pie(feature, label_name):
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('-5 ' + label_name)
    for i in range(3):
        df = (games_actual.groupby(feature)[regions[i]].sum().reset_index().sort_values(regions[i], ascending = False, 
                                                                                           ignore_index = True))
        if len(df) > 5:
            df = df.head(5)
            df = df.append({feature: '', regions[i]: games_actual[regions[i]].sum() - df[regions[i]].sum()}, 
                           ignore_index= True)
        axs[i].pie(df[regions[i]], labels = df[feature], autopct='%1.2f%%', 
                   colors=['#610C9F','#940B92','#DA0C81','#E95793','#DCBFFF','#D0A2F7'], 
                   textprops={'color':"w"}, explode=len(df)*[0.1])
        axs[i].set_title(df.columns[1][:2].upper())
    plt.show()
```

###  5.  

####     Xbox One  PC .

- H0:     Xbox One  PC .
- H1:     Xbox One  PC .

     -1

```python
user_scores = games_actual.query('user_score >= 0')
```

```python
alpha = 0.05
results = st.ttest_ind(user_scores.query('platform == "XOne"')['user_score'], 
                       user_scores.query('platform == "PC"')['user_score'], equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('  ')
else:
    print('    ')
```

####     Action (. , -)  Sports (.  ) .

- H0:     Action  Sports .
- H1:     Action  Sports .

```python
alpha = 0.05
results = st.ttest_ind(user_scores.query('genre == "Action"')['user_score'], 
                       user_scores.query('genre == "Sports"')['user_score'], equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('  ')
else:
    print('    ')
```
