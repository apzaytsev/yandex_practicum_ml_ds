# Разработка моделей машинного обучения для управления рисками и принятия решения о покупке

К нам обратился фермер, владелец молочного хозяйства «Вольный луг», который хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а надой был вкусным — строго по его критериям.

Необходимо разработать модель машинного обучения, которая поможет управлять рисками и принимать объективное решение о покупке. «ЭкоФерма» готова предоставить подробные данные о своих коровах. Нужно создать 2 прогнозные модели для отбора бурёнок в поголовье:
- Первая будет прогнозировать возможный удой у коровы (целевой признак Удой);
- Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).

С помощью модели нужно отобрать коров по двум критериям:
- Средний удой за год — не менее 6000 килограммов,
- Молоко должно быть вкусным.

<b>Целью проекта</b> является разработка моделей для прогнозирования удоя молока и его вкуса с целью отбора подходящих для покупки коров в "ЭкоФерме".

##  Предобработка данных

Проверим данные на наличие пропусков и дубликатов. Проверим корректность типов данных у каждого признака. При необходимости устраним все проблемы с данными.

```python
# Переименуем столбцы для удобства использования.

dict_columns = {'Удой, кг': 'milk', 'ЭКЕ (Энергетическая кормовая единица)': 'eke', 'Сырой протеин, г': 'crude_protein', 
                'СПО (Сахаро-протеиновое соотношение)': 'spo', 'Порода': 'breed', 'Тип пастбища': 'pasture', 
                'порода папы_быка': 'father_breed', 'Жирность,%': 'fat', 'Белок,%': 'protein', 'Вкус молока': 'milk_taste', 
                'Возраст': 'age', 'Имя Папы': 'father_name', 'Имя_папы': 'father_name', 'Текущая_жирность,%': 'fat', 
                'Текущий_уровень_белок,%': 'protein'}
```

```python
# Создадим функцию для получения ключа по значению в дальнейшем для заголовков графиков.

def get_key_from_dict(val):
    for key, value in dict_columns.items():
        if val == value:
            return key
```

```python
df_farm = df_farm.rename(columns=dict_columns)
df_dad = df_dad.rename(columns=dict_columns)
df_new_cows = df_new_cows.rename(columns=dict_columns)
```

```python
# Удалим дубликаты

df_farm = df_farm.drop_duplicates()
df_new_cows = df_new_cows.drop_duplicates()
```

```python
# Далее создадим функцию для приведения к единообразию оформления.

def preprocessing(df):
    for i in df.columns:
        if df[i].dtype == 'object':
            df[i] = df[i].str.lower()
            df[i] = df[i].str.replace('_', '')
            df[i] = df[i].str.replace(' ', '')
```

```python
preprocessing(df_farm)
preprocessing(df_dad)
preprocessing(df_new_cows)
```

```python
# Заменим неявные дубликаты.

df_farm['pasture'] = df_farm['pasture'].str.replace('равнинные', 'равнинное')
df_farm['father_breed'] = df_farm['father_breed'].str.replace('айдиалл', 'айдиал')
```

### Построим графики для каждого признака раздельно для количественных и категориальных признаков.

```python
plt.style.use('cyberpunk')
```

```python
# Создадим функцию для создания графиков вида pie для категориальных признаков и гистограммы - для количественных.

def draw_charts(df, df_label):
    list = []
    list2 = []
    k = 0
    for i in df.columns:
        if df[i].dtype == 'object':
            list.append(i)
        else:
            if i != 'id':
                list2.append(i)
    fig, axs = plt.subplots(len(list)//3, 3, figsize=(18, 5*(len(list)//3)))
    axs = axs.flat
    fig.suptitle('Категориальные признаки ' + df_label)
    for j in range(len(list)//3*3):
        axs[j].set_title(label=get_key_from_dict(list[j]))
        axs[j].pie(df[list[j]].value_counts(), labels=df[list[j]].value_counts().index.values.tolist(), autopct='%1.2f%%', 
                   colors=['#610C9F','#E95793'], textprops={'color':"w"}, 
                   explode=len(df[list[j]].value_counts().index.values.tolist()) * [0.05])
    plt.show()
    if len(list)%3 != 0:
        fig, axs = plt.subplots(1, len(list)%3, figsize=(6*(len(list)%3), 5))
        axs = axs.flat
        for j in range(len(list) - len(list)//3*3):
            j1 = j+len(list)//3*3
            axs[j].set_title(label=get_key_from_dict(list[j1]))
            axs[j].pie(df[list[j1]].value_counts(), labels=df[list[j1]].value_counts().index.values.tolist(), autopct='%1.2f%%', 
                       colors=['#610C9F','#E95793'], textprops={'color':"w"}, 
                       explode=len(df[list[j1]].value_counts().index.values.tolist()) * [0.05])
        plt.show()
    axs = df[list2].hist(bins=100, figsize=(18, 6*(len(list2)//2)))
    for ax in axs.flatten():
        ax.set_title(get_key_from_dict(list2[k]))
        k += 1
    for i in range(len(list2)):
        plt.figure(figsize=(18, 2))
        df.boxplot(column=list2[i], vert=False)
        plt.title(get_key_from_dict(list2[i]))
        plt.show()
```

```python
draw_charts(df_farm, 'данных о стаде фермера на текущий момент')
```

```python
draw_charts(df_new_cows, 'данных о коровах «ЭкоФермы»')
```

## Обучение модели линейной регрессии

### Построение 1й модели линейной регрессии

```python
# Создадим функцию для обучения модели.

model_lr = LinearRegression()
clf = LogisticRegression()

def learn_model(df, dropped_cols, cat_col_names, num_col_names, target, model):
    RANDOM_STATE = 42
    X = df.drop(dropped_cols, axis=1)
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE)

    encoder = OneHotEncoder(drop='first', sparse=False)
    X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
    X_test_ohe = encoder.transform(X_test[cat_col_names])
    encoder_col_names = encoder.get_feature_names()
    X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
    X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train[num_col_names])
    X_test_scaled = scaler.transform(X_test[num_col_names])
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)
    
    X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
    X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
    
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    residuals =  y_test - predictions

    if model == 'model_lr':
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6))
        axes[0].hist(residuals, bins=50)
        axes[0].set_title('Гистограмма распределения остатков')
        axes[0].set_xlabel('Остатки')
        axes[1].scatter(x=predictions, y=residuals)
        axes[1].set_xlabel('Предсказания модели')
        axes[1].set_ylabel('Остатки')
        axes[1].set_title('Анализ дисперсии')
        plt.show()

        r2 = r2_score(y_test, predictions)
        print('R2 =', r2)

    return y_test, predictions, residuals, model, X_train, X_test, encoder, scaler
```

```python
dropped_cols = ['protein', 'fat', 'milk_taste', 'milk', 'id']
cat_col_names = ['age', 'breed', 'pasture', 'father_breed']
num_col_names = ['eke', 'crude_protein', 'spo']

y_test_1, predictions_1, residuals_1, model_1, X_train_1, X_test_1, encoder_1, scaler_1 = learn_model(
    df_farm, dropped_cols, cat_col_names, num_col_names, 'milk', model_lr)
```

Чем ближе значение оценки R² к 1, тем точнее обучается модель. По своей сути это разница между выборками в наборе данных и прогнозами, сделанными моделью. Значение метрики показывает какой процент дисперсии в данных объясняет наша модель. 

Сопоставляя значения R² трех моделей (0.7844078173416966, 0.8180879926867503 и 0.8247542745210406 соответственно) можно сделать вывод, что наилучшее значение метрики имеет 3я модель. 

Графики остатков показывают как в остатках обнаруживаются выбросы, что обуславливает необходимость добавления в 1ю и 2ю модели дополнительных признаков.
Заметим, что гистограмма 3ей модели стремится к нормальному распределению по сравнению с 1 и 2.

### Рассчитаем точность каждой модели по трём метрикам: MSE, MAE, RMSE.

```python
def evaluate_by_metrics(y_test, predictions):
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mse)
    print('MSE =', mse, 
          '\nMAE =', mae, 
          '\nRMSE =', rmse)
```

```python
accuracy = accuracy_score(y_test_log, predictions_log)
recall = recall_score(y_test_log, predictions_log)
precision = precision_score(y_test_log, predictions_log)
cm = confusion_matrix(y_test_log, predictions_log)
print('Accuracy =', accuracy,
      '\nPrecision =', precision,
      '\nRecall =', recall)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.title('Матрица ошибок')
plt.ylabel('True Label')
plt.xlabel('Predicted')
```

1. Модель верно классифицирует <b>62.4%</b> значений (accuracy).
2. <b>61.9%</b> предсказанных положительных случаев действительно являются истинными положительными (precision).
3. Модель способно обнаруживать <b>85.2%</b> положительных случаев из всех истинных положительных (recall).

```python
# Настроим порог классификации.

thresholds = [round(i,2) for i in np.linspace(0.1,1,num=50,endpoint=False)]

columns = ['y_pred_'+str(i) for i in thresholds]
data = pd.DataFrame(columns=['y_valid', 'y_proba']+columns)

y_proba = clf.predict_proba(X_test_log)[:,1]
data['y_valid'] = y_test_log
data['y_proba'] = y_proba

accuracy_list, precision_list, recall_list = [], [], []
for i, col in enumerate(columns):
    threshold = thresholds[i]
    data[col] = data['y_proba'].apply(lambda x: 1 if x >= threshold else 0)
    accuracy_list.append(accuracy_score(y_test_log, data[col]))
    precision_list.append(precision_score(y_test_log, data[col]))
    recall_list.append(recall_score(y_test_log, data[col]))
```

```python
# Построим графики метрик для всех порогов.

plt.figure(figsize=(12, 6))
plt.plot(thresholds, accuracy_list, label= 'accuracy')
plt.plot(thresholds, precision_list, label = 'precision')
plt.plot(thresholds, recall_list, label = 'recall')
plt.legend()
plt.show()
```

- При пороговом значении 0.8 precision максимальна, но очень низкая точность.
- Оптимальным значением с наибольшим accuracy будет порог <b>0.6</b>.
